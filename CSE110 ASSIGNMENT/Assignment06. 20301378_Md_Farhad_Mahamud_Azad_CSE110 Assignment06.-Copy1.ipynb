* If the user does not input ‘+’ or ‘-’ or ’/’ or ‘*’ or ‘%’ in second position of the input, again raise an exception stating “Input does not contain 3 elements/Wrong operator”.
* In case of division, also handle divide by zero exception.

**Input:**\

```python
#to do
#taking input bfrom user
inp = input(" Enter num ,oparetor, and number  :")

#checking the input
try:
    list1= inp.split(' ')
    
    if len(list1) != 3:
        raise Exception('Input does not contain 3 elements')
    
    
    for iii in list1[0::2]:
        if iii.isdigit() != True:
            raise ValueError ('Input does not contain numbers.') 
            
    opa =[ '+', '-' , '*' ,'/' ,"%"]
    
    for jjj in list1[1::2]:
        
        if (jjj not in opa):
            raise Exception('Wrong operator')
        if (jjj=="+"):
            sum1= int(list1[0]) + int(list1[2])
            print('The ans is : ',sum1)
        if (jjj=="-"):
            sub1= int(list1[0]) - int(list1[2])
            print('The ans is : ',sub1)
        if (jjj=="*"):
            mul1= int(list1[0]) * int(list1[2])
            print('The ans is : ',mul1)
        if (jjj=="/"):
            div1= int(list1[0]) / int(list1[2])
            print('The ans is : ',div1)
        if (jjj=="%"):
            mod1= int(list1[0]) % int(list1[2])
            print('The ans is : ',mod1)
            
#handling the error
            
except ValueError as eee:
    print(eee)
except Exception as fff:
    print(fff)

#code ended
```
